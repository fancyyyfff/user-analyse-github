```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime, timedelta
import os

# 设置中文字体（可选）
plt.rcParams['font.sans-serif'] = ['SimHei']  
plt.rcParams['axes.unicode_minus'] = False

# 创建可视化文件夹
os.makedirs('visualizations', exist_ok=True)

# 1. 生成模拟数据并保存为 CSV
print("生成模拟数据...")

n_users = 10000
n_records = 200000
base_date = datetime(2023, 6, 1)

user_ids = np.random.randint(1, n_users + 1, n_records)
event_dates = [base_date + timedelta(days=np.random.randint(0, 30)) for _ in range(n_records)]

df = pd.DataFrame({
    'user_id': user_ids,
    'datetime': event_dates,
    'event_type': np.random.choice(['view', 'cart', 'purchase'], n_records, p=[0.7, 0.2, 0.1]),
    'product_id': np.random.randint(1000, 5000, n_records)
})

# 加入空值
df.loc[np.random.choice(df.index, size=500, replace=False), 'product_id'] = np.nan

# 保存为 CSV
df.to_csv('ecommerce_data.csv', index=False)
print("保存为 ecommerce_data.csv")

# 2. 从 CSV 读取数据
df = pd.read_csv('ecommerce_data.csv', parse_dates=['datetime'])
print("读取数据成功：")
print(df.head())

# 3. 事件类型分布
plt.figure(figsize=(8, 5))
event_counts = df['event_type'].value_counts()
sns.barplot(x=event_counts.index, y=event_counts.values, palette="Set2")
plt.title('事件类型分布')
plt.xlabel('事件类型')
plt.ylabel('数量')
plt.savefig('visualizations/event_type_distribution.png', bbox_inches='tight')
plt.show()
plt.close()

# 4. 每天的事件数量
daily_events = df.groupby(df['datetime'].dt.date).size()
plt.figure(figsize=(12, 6))
daily_events.plot(kind='bar', color='skyblue')
plt.title('每日事件数量')
plt.xlabel('日期')
plt.ylabel('数量')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('visualizations/daily_event_count.png')
plt.show()
plt.close()

# 5. 用户活跃度分布（活跃天数）
df['date'] = df['datetime'].dt.date
active_days = df.groupby('user_id')['date'].nunique()
plt.figure(figsize=(10, 6))
sns.histplot(active_days, bins=30, kde=True, color='orange')
plt.title('用户活跃天数分布')
plt.xlabel('活跃天数')
plt.ylabel('用户数量')
plt.savefig('visualizations/user_active_days.png')
plt.show()
plt.close()

# 6. 每个用户行为次数分布
user_event_counts = df.groupby('user_id').size()
plt.figure(figsize=(10, 6))
sns.histplot(user_event_counts, bins=30, kde=True, color='green')
plt.title('用户行为次数分布')
plt.xlabel('行为次数')
plt.ylabel('用户数量')
plt.savefig('visualizations/user_event_counts.png')
plt.show()
plt.close()

# 7. 每个用户的购买转化率（购买 / 总行为）
user_events = df.groupby(['user_id', 'event_type']).size().unstack(fill_value=0)
user_events['conversion_rate'] = user_events['purchase'] / (user_events.sum(axis=1))
plt.figure(figsize=(10, 6))
sns.histplot(user_events['conversion_rate'].dropna(), bins=30, kde=True, color='purple')
plt.title('用户购买转化率分布')
plt.xlabel('转化率')
plt.ylabel('用户数量')
plt.savefig('visualizations/user_conversion_rate.png')
plt.show()
plt.close()

print("所有图表已保存并显示完毕。")
```
